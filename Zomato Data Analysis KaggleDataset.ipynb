```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('dark_background')
```

```python
df = pd.read_csv('desktop/zomato.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

```python
#dropping duplicates
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.info()
```

```python
#Cleaning Rate Column
```

```python
df['rate'].unique()
```

```python
#Removing "new","-", and '/5' and 'nan or null values from rate column'
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#How many null values are there
df.rate.isnull().sum()
```

```python
#Now handling the null values in rate column with mean
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
# dropping null values
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type','listed_in(city)':'City'}, inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['City'].unique()
```

```python
#Listed city and location, both are ther, let's keep only one
df=df.drop(['City'],axis=1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

```python
# comma in values is restricting the values to become int or float
```

```python
#Removing comma from Cost2Plates Column
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
#Cleaning rest type column
```

```python
df['rest_type'].unique()
```

```python
rest_types = df['rest_type'].value_counts()
rest_types
```

```python
rest_type_lessthan100 = rest_types[rest_types < 1000]
rest_type_lessthan100
```

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan100):
        return 'others'
    else:
        return value
df['rest_type']  = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts()
```

```python
location_lessthan300 = location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
#Cleaning Cuisines Column
```

```python
cuisines = df['cuisines'].value_counts()

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```
